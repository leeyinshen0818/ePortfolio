```python
import csv

def load_ride_data(filename):
    data = []
    with open(filename, mode='r') as file:
        reader = csv.DictReader(file) 
        #read and parse data iteratively  (5 marks)
        for row in reader:
            print(row['ride_id'],row['driver_name'],row['rider_name'],row['city'],row['distance'],row['duration'],row['fare'])
            data.append(row)
    return data

# Write code(s) to load data from actual CSV file (2 marks)
ride_data = load_ride_data('ride.csv')

# Write code(s) to preview data (1 mark)
print(ride_data)
```

```python
# Write your function definition here to compute average fare by city (8 marks)
def average_fare(filename):
# for each city get the total fare of each occurence divide by the number of occurence
    city_data = {}
    
    with open(filename,mode='r') as file:
        reader = csv.DictReader(file)

        for row in reader:
            city = row['city']
            fare = round(float(row['fare']),2)

            # Initialize city entry if not exists
            if city not in city_data:
                city_data[city] = {'total_fare': 0.0, 'count': 0}

            # For every city occurence update total fare and counter
            city_data[city]['total_fare'] += fare
            city_data[city]['count'] += 1

    # Calculate average fare for each city
    average = {}
    for city, data in city_data.items():
        average[city] = round(data['total_fare'] / data['count'],2)

    return average

# Write the code(s) to call your function and pass the data (2 marks)
avg_fare = average_fare('ride.csv')
print(avg_fare)
```
