```python
# Write your function definition here for top 3 drivers by total fare (5 marks)

import csv

def top_drivers_fare(data):
    # Use a dictionary to calculate total fare of driver
    fare_summary = {}
    
    # Calculate total fare per driver
    for ride in data:
        driver = ride['driver_name']
        fare = float(ride['fare'])
        if driver not in fare_summary:
            fare_summary[driver] = 0
        fare_summary[driver] += fare
    
    # Sort drivers to get top 3
    top_drivers = sorted(fare_summary.items(), key=lambda x: x[1], reverse=True)[:3]
    
    # Return the top 3 drivers 
    return [(driver, total_fare) for driver, total_fare in top_drivers]


# Read the CSV file and load the data
with open('ride.csv', 'r') as file:
    reader = csv.DictReader(file)
    ride_data = list(reader)


# Write the code(s) to call your function and pass the driver's summary data (2 marks)
top_drivers = top_drivers_fare(ride_data)
for driver, total_fare in top_drivers:
    print(f"Driver: {driver}, Total Fare: {total_fare:.2f}")
```

```python
# Write your function definition here for average duration and sort cities (8 marks)

import csv

def average_duration_city(data):
    # Dictionary to store the total duration and count of rides for each city
    city_durations = {}

    # Calculate total duration and count of rides for each city
    for ride in data:
        city = ride['city']
        duration = int(ride['duration'])
        
        if city not in city_durations:
            city_durations[city] = {'total_duration': 0, 'ride_count': 0}
        
        city_durations[city]['total_duration'] += duration
        city_durations[city]['ride_count'] += 1

    # Calculate average duration for each city
    city_avg_duration = {
        city: city_durations[city]['total_duration'] / city_durations[city]['ride_count']
        for city in city_durations
    }

    # Sort city from top to buttom
    sorted_city = sorted(city_avg_duration.items(), key=lambda x: x[1], reverse=True)

    return sorted_city


with open('ride.csv', 'r') as file:
    reader = csv.DictReader(file)
    ride_data = list(reader)


# Write the code(s) to call your function and pass the data (2 marks)
sorted_cities = average_duration_city(ride_data)

for city, avg_duration in sorted_cities:
    print(f"City: {city}, Average Duration: {avg_duration:.2f} minutes")
```

```python
# Write your function definition here for writing driver summary into a text file (8 marks)

import csv

def driver_summary_txt(data, filename):
    # Dictionary to store driver summary data
    driver_summary = {}
    
    # Summarize data for each driver
    for ride in data:
        driver = ride['driver_name']
        distance = float(ride['distance'])
        fare = float(ride['fare'])

        if driver not in driver_summary:
            driver_summary[driver] = {'rides': 0, 'total_distance': 0, 'total_fare': 0}

        driver_summary[driver]['rides'] += 1
        driver_summary[driver]['total_distance'] += distance
        driver_summary[driver]['total_fare'] += fare

    # Write the summary to txt file
    with open(filename, 'w') as file:
        # Write header
        file.write("Driver, Total Rides, Total Distance, Total Fare\n")
        
        # Write driver data
        for driver, summary in driver_summary.items():
            file.write(f"{driver}, {summary['rides']}, {summary['total_distance']}, {summary['total_fare']:.2f}\n")


with open('ride.csv', 'r') as file:
    reader = csv.DictReader(file)
    ride_data = list(reader)


# Write the code(s) to call your function and pass the driver summary data (2 marks)
driver_summary_txt(ride_data, 'driver_summary.txt')

```
